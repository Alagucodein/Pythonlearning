```python
df = df.drop('Id',axis=1)
```

#Species Label Encoding

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
```

```python
df1 = df.copy()
```

```python
x = df1.iloc[:,0:4]
```

```python
x
```

```python
y = df1.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=334)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model1 = KNeighborsClassifier()
```

```python
model1.fit(x_train,y_train)
```

```python
y_predict = model1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm1 = confusion_matrix(y_test, y_predict)
```

```python
cm1
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm1,annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
from sklearn.svm import SVC
```

```python
model2 = SVC()
```

```python
model2.fit(x_train, y_train)
```

```python
y_predict1 = model2.predict(x_test)
```

```python
cm2 = confusion_matrix(y_test, y_predict1)
```

```python
cm2
```

```python
print(classification_report(y_test, y_predict1))
```

```python
sns.heatmap(cm2,annot=True)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model3 = LogisticRegression()
```

```python
model3.fit(x_train,y_train)
```

```python
y_predictlr = model3.predict(x_test)
```

```python
cm3 = confusion_matrix(y_test,y_predictlr)
```

```python
cm3
```

```python
print(classification_report(y_test, y_predictlr))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 5))
titles = ["KNN", "SVC", "Logistic Regression"]
cms = [cm1, cm2, cm3]

for ax, cm, title in zip(axes, cms, titles):
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(ax=ax, colorbar=False)
    ax.set_title(title)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Define the three confusion matrices
#cm1 = np.array([[90, 10], [5, 95]])
#cm2 = np.array([[85, 15], [10, 90]])
#cm3 = np.array([[95, 5], [3, 97]])

# Define the labels for the confusion matrices
labels = ['Class 0', 'Class 1']

# Create a figure with three subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot the first confusion matrix
sns.heatmap(cm1, ax=axs[0], annot=True, cmap='Blues', fmt='d')
axs[0].set_title('Confusion Matrix 1')
axs[0].set_xlabel('Predicted Labels')
axs[0].set_ylabel('True Labels')
axs[0].set_xticks([0, 1])
axs[0].set_yticks([0, 1])
axs[0].set_xticklabels(labels)
axs[0].set_yticklabels(labels)

# Plot the second confusion matrix
sns.heatmap(cm2, ax=axs[1], annot=True, cmap='Blues', fmt='d')
axs[1].set_title('Confusion Matrix 2')
axs[1].set_xlabel('Predicted Labels')
axs[1].set_ylabel('True Labels')
axs[1].set_xticks([0, 1])
axs[1].set_yticks([0, 1])
axs[1].set_xticklabels(labels)
axs[1].set_yticklabels(labels)

# Plot the third confusion matrix
sns.heatmap(cm3, ax=axs[2], annot=True, cmap='Blues', fmt='d')
axs[2].set_title('Confusion Matrix 3')
axs[2].set_xlabel('Predicted Labels')
axs[2].set_ylabel('True Labels')
axs[2].set_xticks([0, 1])
axs[2].set_yticks([0, 1])
axs[2].set_xticklabels(labels)
axs[2].set_yticklabels(labels)

# Layout so plots do not overlap
fig.tight_layout()

plt.show()
```
