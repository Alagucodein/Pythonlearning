x,y split .....

```python
x = df[['X']]
```

```python
y = df[['Y']]
```

```python
y
```

train test split ........

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1234)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_predict = lr.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,y_predict)
```

```python
MAE
```

```python
MSE = mean_squared_error(y_test,y_predict)
```

```python
MSE
```

```python
RMSE = np.sqrt(MSE)
```

```python
RMSE
```

```python
plt.scatter(x,y)
plt.plot(x, lr.predict(x), color='red')
plt.title('Regression Table (Linear Regression)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) # Tranforming
X_poly = poly_reg.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x, lin_reg2.predict(poly_reg.fit_transform(x)), color='red')
plt.title('Regression Table (Polynomial Regression)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
