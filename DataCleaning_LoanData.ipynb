```python
df2 = df2.dropna(subset=['Married'])
```

```python
a=['Gender','Dependents','Self_Employed']
```

```python
df2[a] = df2[a].fillna(df2.mode().iloc[0])
```

```python
df2.isnull().sum()
```

```python
b=['LoanAmount','Loan_Amount_Term','Credit_History']
```

```python
df2[b] = df2[b].fillna(df2.min())
```
