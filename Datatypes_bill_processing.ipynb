```python
icecream = 10
chocolate = 5
cake = 15
a = int(input('No of icecream :'))
b = int(input('No of chocolate :'))
c = int(input('No of cake :'))
total = (a * icecream) + (b * chocolate) + (c * cake)
print('icecream : Rs', icecream ,'* Quantity', a , '=', (icecream * a), ',', 'chocolate : Rs', chocolate ,'* Quantity', b ,'=' ,(chocolate * a), ',', 'cake : Rs', cake, '* Quantity' ,c ,'=' ,(cake*c),'Total = ', total)
```
